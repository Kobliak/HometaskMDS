```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

** Уровень 0 **

```python
def f(x):
    return x**5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)
```

```python
f_1 = derivative(f, 1.0, dx=1e-6)
f_2 = derivative(f, 1.0, dx=1e-6, n=2, order=3)
```

** Уровень 1 **

```python
def g(x):
    return ((np.sin(2*x + 1))**5)
```

```python
x = np.linspace(-5, 5)
gx = g(x)
g1x = [derivative(g, xx, dx=1e-6, n=1) for xx in x]
g2x = [derivative(g, xx, dx=1e-6, n=2, order=3) for xx in x]
```

```python
plt.plot(x, gx, label='func')
plt.plot(x,g1x, label = '1st der')
plt.plot(x,g2x, label = '2st der')
plt.legend()
plt.show()
```
