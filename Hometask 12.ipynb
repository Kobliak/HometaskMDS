```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import Lasso, Ridge, MultiTaskLasso, HuberRegressor, ElasticNet, SGDRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
boston = load_boston()
X = pd.DataFrame(boston.data)
X.columns = boston.feature_names
y=pd.DataFrame(boston.target)
y.columns=['MEDV']
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_valid = sc.transform(X_valid)
```

```python
models=[
    {'name':'Ridge',"model": Ridge(), 'params':{
                                                'alpha': np.logspace(-7, 2, 1000), 
                                                'solver': ['svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
                                                }},
    {'name':'Lasso',"model":Lasso(), 'params':{'alpha': np.logspace(-7, 2, 1000)}},  
    {'name':'MTLasso',"model": MultiTaskLasso(), 'params':{
                                                           'alpha': np.logspace(-7, 2, 1000), 
                                                            'selection': ['cyclic', 'random']
                                                           }},
    {'name':'Elastic',"model": ElasticNet(), 'params':{
                                                  'alpha': np.logspace(-7, 2, 1000), 
                                                  'l1_ratio': np.linspace(0, 1, 50),
                                                  'selection': ['cyclic', 'random']
                                                  }},
    {'name':'SGDR',"model": SGDRegressor(), 'params':{
                                                  'alpha': np.logspace(-7, 2, 1000), 
                                                  'l1_ratio': np.linspace(0, 1, 50),
                                                  'penalty': ['l2', 'l1', 'elasticnet']
                                                  }},
    {'name':'Huber',"model": HuberRegressor(), 'params':{
                                                  'alpha': np.logspace(-7, 2, 1000), 
                                                  'epsilon': np.linspace(1.35, 2, 50)
                                                  }},
    {'name':'DTreeR',"model": DecisionTreeRegressor(), 'params':{
                                                                'max_depth': range(1, 11),
                                                                'splitter': ['best', 'random'],
                                                                'criterion': ['mse', 'mae', 'friedman_mse'],
                                                                'min_samples_leaf': [1, 2, 4, 8, 16]
                                                  }}
]
```

```python
%%time

res=[]
for v in  models:
    res.append((v['name'], RandomizedSearchCV(v['model'], v['params'], cv=10).fit(X_train, y_train)))
```

```python
for r in res:
    print(r[0], 'Score train: ', r[1].best_score_, 'Score valid: ', r[1].best_estimator_.score(X_valid, y_valid))
```

```python
for r in res:
    print(r[0], r[1].best_params_)
```
