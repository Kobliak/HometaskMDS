```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
lam = 0.1
rv = sts.expon(lam)
sample = rv.rvs(1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
n02 = 2
n05 = 5
n10 = 10
n30 = 30
n50 = 50
n100 = 1000

samples_count = 1000

a02 = np.array([])
a05 = np.array([])
a10 = np.array([])
a30 = np.array([])
a50 = np.array([])
a100 = np.array([])

for number in range(samples_count):
    # на каждом цикле генерируем выборки разных размеров
    sample02 = rv.rvs(size=n02)
    sample05 = rv.rvs(size=n05)
#    sample10 = rv.rvs(size=n10)
    sample30 = rv.rvs(size=n30)
    sample50 = rv.rvs(size=n50)
    sample100 = rv.rvs(size=n100)
    # считаем среднее по каждой выборке и добавляем в массив средних
    a02 = np.append(a02, sample02.mean())
    a05 = np.append(a05, sample05.mean())
#    a10 = np.append(a10, sample10.mean())
    a30 = np.append(a30, sample30.mean())
    a50 = np.append(a50, sample50.mean())
    a100 = np.append(a100, sample100.mean())
```

```python
#строим графики
#подписи для графиков
label_1 = 'sample of ' + str(n02) + ' histogram'
label_2 = 'sample of ' + str(n05) + ' histogram'
label_3 = 'sample of ' + str(n10) + ' histogram'
label_4 = 'sample of ' + str(n30) + ' histogram'
label_5 = 'sample of ' + str(n50) + ' histogram'
label_6 = 'sample of ' + str(n100) + ' histogram'

#plt.hist(a02, density=True, label=label_1)
plt.hist(a05, density=True, label=label_2)
#plt.hist(a10, density=True, label=label_3)
plt.hist(a30, density=True, label=label_4)
plt.hist(a50, density=True, label=label_5)
plt.hist(a100, density=True, label=label_6)


plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```
