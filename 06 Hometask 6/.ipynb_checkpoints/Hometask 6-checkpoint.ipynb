```python
import numpy as np
from scipy.optimize import differential_evolution
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]
```

```python
#### TOL
TOL = [0.01, 0.1, 0.3, 0.4]
RES_TOL = []
for tol in TOL:
    result = differential_evolution(ackley, bounds, tol=tol, seed=42)
    RES_TOL.append(result.nit)
    
plt.plot(TOL, RES_TOL)
plt.title("Количество итераций в зависимости от TOL")
plt.xlabel("TOL")
plt.ylabel("Количество итераций")
plt.grid()
```

```python
###### MUTATION
MUTATION = np.linspace(0, 1.999, 10)
RES_MUT = []
for mutation in MUTATION:
    result = differential_evolution(ackley, bounds, strategy='best1bin', tol=0.01, mutation=mutation, seed=42)
    RES_MUT.append(result.nit)

x = MUTATION
y = RES_MUT

fig, ax = plt.subplots()
ax.bar(x, y)
ax.set_title('Количество итераций в зависимости от MUTATION в [0,2) ')
ax.set_xlabel('MUTATION')
ax.set_ylabel('Количество итераций')
plt.show()
```

```python
## Recombination 

RECOMBINATION = np.linspace(0, 1, 10)
RES_RECOMB = []
for recombination in RECOMBINATION:
    result = differential_evolution(ackley, bounds, strategy='best1bin', tol=0.01, recombination=recombination, seed=42)
    RES_RECOMB.append(result.nit)

plt.plot(RECOMBINATION, RES_RECOMB)
plt.title("Количество итераций в зависимости от RECOMBINATION в [0,1]")
plt.xlabel("RECOMBINATION")
plt.ylabel("Количество итераций")
plt.grid()
```

```python
### UPDATING
UPDATING = ['immediate', 'deferred']
RES_UPDATING = []
for updating in UPDATING:
    result = differential_evolution(ackley, bounds, strategy='best1bin', tol=0.01, updating=updating, seed=42)
    RES_UPDATING.append(result.nit)

x = UPDATING
y = RES_UPDATING

fig, ax = plt.subplots()
ax.bar(x, y)
ax.set_title('Количество итераций в зависимости от UPDATING')
ax.set_xlabel('UPDATING Type')
ax.set_ylabel('Количество итераций')
plt.show()
```

```python
POPSIZE = [5, 10, 13, 15, 17, 20, 30, 40]
RES_POPSIZE = []
for popsize in POPSIZE:
    result = differential_evolution(ackley, bounds, popsize=popsize, seed=42)
    RES_POPSIZE.append(result.nit)

plt.plot(POPSIZE, RES_POPSIZE)
plt.title("Количество итераций в зависимости от POPSIZE")
plt.xlabel("POPSIZE")
plt.ylabel("Количество итераций")
plt.grid()
```

```python
STRATEGY = ['best1bin', 'best1exp', 'rand1exp', 'randtobest1exp', 'currenttobest1exp', 'best2exp', 'rand2exp',
            'randtobest1bin', 'currenttobest1bin', 'best2bin', 'rand2bin', 'rand1bin']
res_strategy = []
for strategy in STRATEGY:
    result = differential_evolution(ackley, bounds, strategy=strategy, seed=42)
    res_strategy.append(result.nit)
    
plt.plot(STRATEGY, res_strategy)
plt.title("Количество итераций в зависимости от STRATEGY")
plt.xlabel("STRATEGY")
plt.ylabel("Количество итераций")
plt.grid()
```
